```python
def convert_publish_date(publish_str):
    match = re.search(r'(\w+\s\d{1,2},\s\d{4})', publish_str)
    if match:
        try:
            date = pd.to_datetime(match.group(1))
            return date.strftime('%d/%m/%Y')  # Use '%Y' for the full year instead of '%y'
        except pd.errors.OutOfBoundsDatetime:
            return None
    return None
```

```python
# Áp dụng hàm chuyển đổi cho cột 'publish'
import re

df['publish'] = df['publish'].apply(convert_publish_date)
```

```python
# Xóa ký hiệu phần trăm và các ký tự không cần thiết trong các cột cụ thể
percentage_columns = ['fivestars', 'fourstars', 'threestars', 'twostars', 'onestar']
for col in percentage_columns:
    df[col] = df[col].str.replace('%', '', regex=False)  # Xóa ký hiệu phần trăm
    df[col] = df[col].str.replace(r'[^0-9]', '', regex=True)  # Xóa tất cả ký tự không phải số
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)  # Chuyển đổi sang kiểu số và thay NaN bằng 0
```

```python
# Tạo cột 'cover' từ cột 'pages'
df['cover'] = df['pages'].str.extract(r'(\bHardcover\b|\bPaperback\b)')  # Tìm kiếm 'Hardcover' hoặc 'Paperback'
df['pages'] = pd.to_numeric(df['pages'].str.extract(r'(\d+)')[0], errors='coerce').fillna(0)  # Lấy số trang và chuyển sang kiểu số
```

```python
# Tách cột 'price' và 'category'
df[['category', 'price']] = df['price'].str.extract(r'([A-Za-z\s]+)\s*\$?([\d,\.]+)', expand=True)
df['category'] = df['category'].fillna('Unknown')
df['price'] = df['price'].str.replace(',', '', regex=False)
df['price'] = pd.to_numeric(df['price'], errors='coerce').fillna(0)
```
